```python
import sys
sys.setrecursionlimit(10**6)

def solution(A, B):
    while A != 0 and A%10!=1:
        if (A%10)!= 0 or B==1:
            print(A%10, end='')
            B = 1
        
        solution(A//10, B)

solution(123456789, 0)
```

```python
import sys 

sys.setrecursionlimit(10**6)

n = int(input())

def solution(n, ans):
    ans += n
    while n > 0:
        solution(n-1, ans)
    return ans 

solution(n, 0)
```

```python
import sys 

sys.setrecursionlimit(10**6)

n = int(input())

def solution(n):
    if n == 0:
        return 0 
    return n + solution(n-1)

solution(n)
```
