```python
# 입력값 
nums = [2, 7, 8, 11]
target = 9

hashtable_dict = {} # dict

for i in range(0, len(nums)):
    print(f'i : {i}')
    val = target - nums[i]
    print(f'target, nums[i], val : {target, nums[i], val}')
    # if hashtable_dict.get(val) and ~:
    if hashtable_dict.get(val) is not None and hashtable_dict[val] != i:
        print(sorted([i, hashtable_dict[val]]))
    
    hashtable_dict[nums[i]] = i 

print([-1, -1])
```
