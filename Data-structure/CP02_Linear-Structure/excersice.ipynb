```python
import sys 

sys.setrecursionlimit(10**6)

n = int(input())

def solution(n, ans):
    ans += n
    while n > 0:
        solution(n-1, ans)
    return ans 

solution(n, 0)
```

```python
def is_palindrome(input_str):
    s,e = 0, len(input_str)-1 
    while s < e:
        if input_str[s]!=input_str[e]:
            return False 
        else:
            s += 1 
            e -= 1
            if input_str[s]!=input_str[e]:
                return False 

    return True 

while True:
    string = input().strip()
    if string.startswith('0'):
        break 
    else:
        if is_palindrome(string): print('yes')
        else: print('no')
```
