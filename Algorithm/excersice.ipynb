```python
n = int(input())
n
```

```python
# dp[i][j] => i자릿수의 마지막 자릿수 값이 j일 때 가질 수 있는 경우의 수
dp = list([0]*10 for _ in range(n+1))
dp
```

```python
for j in range(1, 10):
    dp[1][j] = 1
dp
```

```python
for i in range(2, n + 1):
    for j in range(1, 10):
        s = max(j - 2, 1)
        e = min(j + 2, 9)

        for k in range(s, e + 1):
            dp[i][j] += dp[i - 1][k]

        dp[i][j] %= 987654321
```
