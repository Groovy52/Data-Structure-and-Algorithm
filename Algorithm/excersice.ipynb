```python

def solution(A, n):
    l = len(A)
    if l == n:
        return 1 
    if l == 0:
        s, e = 1, 9 
    else:
        s = max(A[l-1] - 2, 1)
        e = min(A[l-1] + 2, 9)

    count = 0
    for i in range(s, e + 1):
        A.append(i)
        count += solution(A, n)
        A.pop()

    return count 

solution([], n)
```

```python
n = int(input())

dic = dict()
for _ in range(n):
    age, name = input().split()
    if int(age) in dic:
        dic[int(age)].append(name)
    else:
        dic[int(age)] = [name] 

for item in sorted(dic.items()):
    if len(item[1]) >= 2:
        print(item[0], item[1])
    print(item[0], item[1])
```

```python
n = int(input())
coord = []

for _ in range(n):
    x, y = map(int, input().split())
    coord.append((x, y))

coord.sort(key = lambda x: (x[0], x[1]))

for x, y in coord:
    print(x, y)
```

```python
def binary_search(target, data):
    start = 0
    end = len(data)-1
    while start <= end:
        mid = (start + end) // 2

        if data[mid] == target:
            return True 
        elif data[mid] > target:
            end = mid -1 
        else:
            start = mid + 1
    return 
```

```python
binary_search(1, N_li)
```

```python
for n in M_li:
    if binary_search(n, N_li):
        print(1)
    else:
        print(0)
```
