```python
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
    
    def __repr__(self):
        return str(self.data)

class BinarySearchTree:
    def __init__(self):
        self.__root = None 
    
    def insert(self, data):
        print('insert! =================', data)
        print(self.__root, data)
        self.__root = self._insert_rec(self.__root, data)
        print('after 1================')
        print(self.__root)

    def _insert_rec(self, node, data):
        if not node:
            node = Node(data)
            print('1) ================')
            print(node)
        else:
            if node.data > data:
                print('2) ================')
                print(node.data, data)
                print(node.left, data)
                node.left = self._insert_rec(node.left, data)
                print('after 2) ================')
                print(node.data, node.left)
            else:
                print('3) ================')
                print(node.data, data)
                print(node.right, data)
                node.right = self._insert_rec(node.right, data)
                print('after 3) ================')
                print(node.data, node.right)

        return node
```

```python
bst = BinarySearchTree()
bst.insert(20)
bst.insert(25)
bst.insert(23)
```
