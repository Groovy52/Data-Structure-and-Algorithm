```python
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
    
    def __repr__(self):
        return str(self.data)

class BinarySearchTree:
    def __init__(self):
        self.__root = None 
    
    def insert(self, data):
        print('insert! =================', data)
        print(self.__root, data)
        self.__root = self._insert_rec(self.__root, data)
        print('after 1================')
        print(self.__root)

    def _insert_rec(self, node, data):
        if not node:
            node = Node(data)
            print('1) ================')
            print(node)
        else:
            if node.data > data:
                print('2) ================')
                print(node.data, data)
                print(node.left, data)
                node.left = self._insert_rec(node.left, data)
                print('after 2) ================')
                print(node.data, node.left)
            else:
                print('3) ================')
                print(node.data, data)
                print(node.right, data)
                node.right = self._insert_rec(node.right, data)
                print('after 3) ================')
                print(node.data, node.right)

        return node
```

```python
bst = BinarySearchTree()
bst.insert(20)
bst.insert(25)
bst.insert(23)
```

```python
n, k = map(int, input().split())
edges = list(list(map(int, input().split())) for _ in range(n-1))
values = list(map(int, input().split()))

def solution(n, k, edges, values):
    E = list([] for _ in range(n))
    for p, c in edges:
        E[p].append(c)

    return tree_depth(0, 0, k, values, E)

def tree_depth(curr_n, curr_d, k, V, E):
    if V[curr_n] == k:
        return curr_d 
    
    for next_n in E[curr_n]:
        ret = tree_depth(next_n, curr_d + 1, k, V, E)
        if ret != -1:
            return ret 
    return -1

print(solution(n, k, edges, values))
```
